```python
questions = ["1. How many people have diabetes and how many people doesn't have diabetes?"
             "2. People of which sex has most diabetes?"
             "3. How does Body Mass Index correlate with the presence of diabetes?"
             "4. What is the distribution of HbA1c levels among diabetic patients?"
             "5. How does medication adherence differ between diabetic and non-diabetic patients?"]

questions
```

```python
#Let's find the answer of first question.

#1. How many people have diabetes and how many people doesn't have diabetes?

df.Diagnosis.value_counts()
```

```python
#plotting bar chart
df.Diagnosis.value_counts().plot(kind = 'bar', color=["orchid", "lightskyblue"])
plt.title("Diabetes Values")
plt.xlabel("1 = Diabetes, 0 = No Diabetes")
plt.ylabel("Amount");
```

```python
df.Diagnosis.value_counts().plot(kind = 'pie', figsize = (8, 6))
plt.legend(["Diabetes", "No diabetes"]);
```

```python
df.Gender.value_counts()
```

```python
#'0' represent 'Male'

#'1' represent 'Female'

#GENDER column part
df.Gender.value_counts().plot(kind = 'bar', figsize = (8, 6), color=["springgreen", "salmon"])
plt.title('Male Female ratio')
plt.legend(['Male', 'Female']);
```

```python
#Let's find the answer of our second question

#2. People of which sex has most diabetes?

pd.crosstab(df.Diagnosis, df.Gender)
```

```python
sns.countplot(x = 'Diagnosis', data= df, hue = 'Gender')
plt.title("Diabetes Frequency for Sex")
plt.xlabel("1 = No Diabetes, 0 = Diabetes")
```

```python
#Let's find the answer of third question.

#3. How does Body Mass Index correlate with the presence of diabetes?

plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='BMI', y='Age', hue='Diagnosis')
plt.title('BMI vs Age with Diabetes Diagnosis')
plt.xlabel('BMI')
plt.ylabel('Age')
plt.show()
```

```python
#Let's find the answer of fourth question.

#4. What is the distribution of HbA1c levels among diabetic patients?

plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='HbA1c', hue='Diagnosis', kde=True)
plt.title('HbA1c Levels by Diabetes Diagnosis')
plt.xlabel('HbA1c Level')
plt.ylabel('Count')
plt.show()
```

```python
#Let's find the answer of fifth question.

#5. How does medication adherence differ between diabetic and non-diabetic patients?

plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='Diagnosis', y='MedicationAdherence')
plt.title('Medication Adherence by Diabetes Diagnosis')
plt.xlabel('Diabetes Diagnosis')
plt.ylabel('Medication Adherence Score')
plt.show()
```
